```python
# dbutils.widgets.text('file_name','')
file_name = dbutils.widgets.get('file_name')
```

```python
from pyspark.sql.functions import current_timestamp, lit

df = spark.read.csv(f'/Volumes/test_catalog/default/test_volume/{file_name}.csv',header=True,inferSchema=True)

final_df = df.withColumn('load_dt',lit(current_timestamp()))
display(final_df)
```

```python
final_df.write.mode('overwrite').saveAsTable(f'default.{file_name}')
```
